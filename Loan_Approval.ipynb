```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
loan_data = pd.read_csv('/content/loan_prediction.csv')
```

```python
loan_data.head()
```

credit history = 1: means they meet the guidelines of the company
1 => good History
0 => bad History
property belonging to them is in urban or rural or semiurban

```python
loan_data.info()
```

Data preprocessing

```python
# check for missing values
loan_data.isnull().sum()
```

out of 614 values these are missing
since the values are quite small  when compared to 614 and some values have categorical value we can drop the missing values

```python
loan_data = loan_data.dropna()
```

```python

loan_data.isnull().sum()
```

```python
loan_data.replace({'Loan_Status': {'Y': 1, 'N': 0}}, inplace = True)
```

```python
loan_data.head()
```

Checking the values in some column

```python
loan_data['Dependents'].value_counts()
```

```python
# we need to convert 3+ to 4
loan_data = loan_data.replace(to_replace='3+', value=4)
loan_data['Dependents'].value_counts()
```

```python
loan_data['Education'].value_counts()
```

```python
loan_data['Self_Employed'].value_counts()
```

```python
loan_data['Credit_History'].value_counts()
```

```python
loan_data['Property_Area'].value_counts()
```

convert the categorial values to numerical values in each column

```python
loan_data.replace({
    'Education': {'Graduate': 1, 'Not Graduate': 0},
    'Self_Employed': {'Yes': 1, 'No': 0},
    'Married': {'Yes': 1, 'No': 0},
    'Gender': {'Male': 1, 'Female': 0},
    },inplace = True)
```

```python
loan_data.head()
```

```python
loan_data.replace({'Property_Area': {'Rural':0,'Semiurban':1,'Urban':2}}, inplace = True)
```

```python
loan_data.head()
```

Extracting essential data

```python
X = loan_data.drop(['Loan_ID','Loan_Status'],axis=1)
Y = loan_data['Loan_Status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting into train and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

Trainning the model

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_prediction = logistic_model.predict(X_train)
```

```python
train_accuracy = accuracy_score(Y_train,X_train_prediction)
print(train_accuracy)
```

```python
X_test_prediction = logistic_model.predict(X_test)
```

```python
test_accuracy = accuracy_score(Y_test,X_test_prediction)
print(test_accuracy)
```

Building a Predictive System

```python
feature_names = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area']

# Create a DataFrame with the same feature names
input_data = (1,1,0,1,1, 3000, 0, 66, 360, 1, 2)
input_data_df = pd.DataFrame([input_data], columns=feature_names)

# Predict
prediction = logistic_model.predict(input_data_df)
print(prediction)
if(prediction):
  print("Loan Accepted")
```
